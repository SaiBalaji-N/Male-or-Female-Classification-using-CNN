```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image
```

```python
import os
import imghdr
import cv2
path=Path("C:/Users/bsai3/Downloads/male or female")
type=["jpg","png","jpeg","bmp"]
for i in os.listdir(path):
    for j in os.listdir(os.path.join(path,i)):
        img_path=os.path.join(path,i,j)
        try:
            image=cv2.imread(img_path)
            img_type=imghdr.what(img_path)
            if img_type not in type:
                os.remove(img_path)
        except Exception as e:
            print("issue with image {}".format(img_path))
```

```python
image_paths={"male":list(path.glob("male images/*")),"female":list(path.glob("female images/*"))}
image_paths["male"][:11]
```

```python
images=[]
for i,j in image_paths.items():
    for k in j:
        image=Image.open(k)
        images.append(image)
images[2]
```

```python
gender={"male":1,"female":0}
```

```python
input=[]
output=[]
import cv2
for i,j in image_paths.items():
    for k in j:
        image=cv2.imread(str(k))
        resized_img=cv2.resize(image,(256,256))
        input.append(resized_img)
        output.append(gender[i])
input
```

```python
output
```

```python
input=np.array(input)
output=np.array(output)
```

```python
from sklearn.model_selection import train_test_split
import tensorflow
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import Sequential
```

```python
cnn=Sequential([
    layers.Conv2D(20,3,padding="same",activation="linear"),
    layers.MaxPooling2D(),
    layers.Conv2D(64,3,padding="same",activation="relu"),
    layers.MaxPooling2D(),
    layers.Conv2D(128,3,padding="same",activation="relu"),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(1,activation="sigmoid")
])
cnn.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
x_train,x_test,y_train,y_test=train_test_split(input,output,test_size=0.2)
```

```python
x_train=x_train.reshape((-1,)+(256,256,3))
x_test=x_test.reshape((-1,)+(256,256,3))
x_train=x_train/255
x_test=x_test/255
```

```python
cnn.fit(x_train,y_train,epochs=100)
```

```python
cnn.predict(x_test)
```

```python
def image(x):
    img=cv2.imread(str(x))
    resized=cv2.resize(img,(256,256))
    resized=np.array(resized)
    reshaped=resized.reshape((-1,)+(256,256,3))
    reshaped=reshaped/255
    predicted=cnn.predict(reshaped)
    if(predicted>0.5):
        plt.title("its a male photo")
        plt.imshow(reshaped.reshape(256,256,3))
        plt.show()
    else:
        plt.title("its a female photo")
        plt.imshow(reshaped.reshape(256,256,3))
        plt.show()
image("C:/Users/bsai3/Downloads/test-2.jpeg")
```
